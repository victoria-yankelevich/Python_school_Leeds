Q2. Try a simple `.plot()` on your temperature data array, to see what xarray does. Why has it done this?

```python
dtemp.plot()
```

Q3. Next, try the same with a 2-dimensional view of your dataset. Try selecting sea surface temperature values and plotting those.

```python
dtemp.squeeze(drop=True).sel(depth=0, method='nearest').plot()
```

```python
dtemp.squeeze(drop=True).sel(depth=200, method='nearest').plot()
```

```python
dtemp.squeeze(drop=True).sel(latitude=0, method='nearest').plot()
```

```python
dtemp.squeeze(drop=True).sel(longitude=0, method='nearest').plot()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
dtemp.squeeze(drop=True).sel(longitude=0, latitude=0, method='nearest').plot(y='depth', ax=ax)
ax.invert_yaxis()
```
